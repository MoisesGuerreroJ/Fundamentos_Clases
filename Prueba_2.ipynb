```python
import numpy as np
import pandas as pd
```

```python
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
df_exam_data[['name', 'score']]
```

#### 2. Se realiza el filtrado tomando df_exam_data.attempts > 2

## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
#1. Se importan los datos desde el repositorio
df_athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
df_athletes
```

```python
#Se filtran los datos obteniendo solo los atletas que han ganado al menos una medalla
df_athletes.loc[(df_athletes['gold'] != 0) | (df_athletes['silver'] != 0) | (df_athletes['bronze'] != 0 )]
```

```python
def muscle_mass(row, weight:str, height:str):
    """
    Calcula el índice de masa muscular (IMC) utilizando la fórmula IMC = row[weight] / row[height]^2
    para un row indicado.

    Parámetros:
    ----------
    row: Indica la fila del DataFrame
    weight (str): Nombre de la columna asociada al peso en kilogramos.
    height (str): Nombre de la columna asociada a la altura en metros.

    Retorna:
    ----------
    float: El valor del índice de masa muscular (IMC) calculado.
    """
    I = row[weight] / (row[height] ** 2)
    return I

df_athletes['muscle_mass'] = df_athletes.apply(muscle_mass, axis = 1, args = ('weight', 'height'))
df_athletes
```

```python
#Se agrupa por deporte y se saca la media de las demás características
df_grouped = df_athletes.groupby('sport', as_index=False)[['height', 'weight', 'gold', 'silver', 'bronze', 'muscle_mass']].mean()

#Se filtran los datos para obtener el pais con el mayor peso promedio de los participantes
df_max_weight = df_grouped[df_grouped['weight'] == df_grouped['weight'].max()]

print(f"El deporte en el que los jugadores tienen mayor peso promedio es '{df_max_weight.sport.values[0]}'"\
        f" con un valor de {df_max_weight.weight.values[0]} kg \n")
print("Datos agrupados por deporte y la media de las características:")
df_grouped
```

```python
#¿Cuál es el pais con más participación femenina?

#Agrupamos por nacionalidad y sexo
df_sex_participation = df_athletes.groupby(['nationality','sex']).size().reset_index(name='count')

#Filtramos por sexo femenino
female_counts = df_sex_participation[df_sex_participation['sex'] == 'female']

#Filtramos para encontrar el país con mayor participacion femenina
country_female_participants = female_counts[female_counts['count'] == female_counts['count'].max()]['nationality'].values[0]

#Imprimimos el resultado
print("El país con la mayor participación femenina es:", country_female_participants)
```

## **Ejercicio 3 (30%)**

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información. Muestre un gráfico de barras con el número de jugadores por grupo de edad diferenciando en esta (mediante un código de color) la posición.
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Punto 1
# Se importan los datos desde un archivo alojado
df = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv')
# Se borran los registros con valores faltantes
df.dropna(inplace=True)
# Se agrupan los datos por la columna Team y se agregan los datos por la suma de Salarios, se organizan descendentemente por este atributo
df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
```

```python
# Punto 2
# Se genera una columna 'Age-Group', en la que se clasifican los jugadores, según su edad, en tres grupos:
# 'Rookie-Sophomore' los menores de 21 años, 'Junior-Senior' entre los 21 y los 32 años, y 'Veteran' para los mayores de 32 años
df['Age-Group'] = pd.cut(df.Age, [0,21,32,40], labels=['Rookie-Sophomore', 'Junior-Senior', 'Veteran'])
df
```

```python
# Se genera un nuevo dataframe con el conteo de jugadores por posición y rango de edad
data = df.groupby(['Age-Group', 'Position'])[['Position']].agg('count')
# Se convierte el dataframe con la función unstack para graficarlo a través de un gráfico de barras
data.unstack().plot(kind='bar', figsize=(12,8), grid=True, title='Cantidad de jugadores por grupo de edad y posición', ylabel='Cantidad', xlabel="Grupo de edad");
```

```python
# Punto 3
# Del punto 1 se infiere que los equipos que pagan mejores salarios son
# Los Angeles Clippers, Oklahoma City Thunder, Golden State Warriors, Cleveland
# Cavaliers, Chicago Bulls
top_high_salaries_teams = ['Los Angeles Clippers', 'Oklahoma City Thunder',
                           'Golden State Warriors', 'Cleveland Cavaliers',
                           'Chicago Bulls']
# Se selecciona del dataframe solo los registros en los eque los equipos estén
# en el listado de los equipos con mejores salarios y se genera la tabla
# dinámica, el índice es la posición, las columnas son los equipos y los valores
# son los salarios promediados
pivot = df[df['Team'].isin(top_high_salaries_teams)].pivot_table(
    index=['Position'], values=['Salary'], columns='Team', aggfunc=['mean'])
# Se aplica un formato de moneda los valores
pivot.applymap('{:,.2f} USD'.format)
```
